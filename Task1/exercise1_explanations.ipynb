### Exercise 1 Explanations - how to use quiver function
For further read: https://www.geeksforgeeks.org/quiver-plot-in-matplotlib/

```python
# Creating one arrow

#starting positon
x_pos = 0
y_pos = 0

#direction vector - define a vector starting from (0,0) and ending in x_direct, y_direct, the direction of the arrow is the direction of this vector  
x_direct = 1
y_direct = 1
 
# Creating plot
fig, ax = plt.subplots(figsize = (12, 7))
ax.quiver(x_pos, y_pos, x_direct, y_direct)
ax.set_title('Quiver plot with one arrow')
 
# Show plot
plt.show()
```

```python
# Creating two arrows, first one starting from (0,0) and with direction (1,1), second one stating from (3,2) and with direction (0,-1)
x_pos = [0, 3]
y_pos = [0, 2]
x_direct = [1, 0]
y_direct = [1, -1]
 
# Creating plot
fig, ax = plt.subplots(figsize = (12, 7))
ax.quiver(x_pos, y_pos, x_direct, y_direct)
 
# show plot
plt.show()
```
